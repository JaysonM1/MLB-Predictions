```python
init = {"Astros": "HOU",
        "Dodgers": "LAD",
        "Rays": "TBR",
        "White Sox": "CHW",
        "Giants": "SFG",
        "Blue Jays": "TOR",
        "Padres": "SDP",
        "Red Sox": "BOS",
        "Athletics": "OAK",
        "Cubs": "CHC",
        "Mets": "NYM",
        "Yankees": "NYY",
        "Braves": "ATL",
        "Phillies": "PHI",
        "Marlins": "MIA",
        "Brewers": "MIL",
        "Angels": "LAA",
        "Reds": "CIN",
        "Indians": "CLE",
        "Nationals": "WSN",
        "Rangers": "TEX",
        "Cardinals": "STL",
        "Mariners": "SEA",
        "Royals": "KCR",
        "Rockies": "COL",
        "Orioles": "BAL",
        "Tigers": "DET",
        "Twins": "MIN",
        "D-backs": "ARI",
        "Pirates": "PIT"
       }

```

```python
def normalize(x, x_min, x_max):
    return (x -x_min)/(x_max-x_min)
def pitching(A, x, era, y, whip, z, so9):
    return A*(x*(1-era) + y*(1-whip) + z* so9)
def batting_form(A, x, ba, y, obp, z, slg):
    return A*(x*ba + y*obp + z * slg)
```

```python
##test factors
A = .6
x1 = .5
y1= .4
z1 = .1
B = .2
x2 = .5
y2 = .3
z2 = .2
C = .2
x3 = .4
y3 = .3
z3 = .3
```

```python
import numpy as np
##testing on random day



rate1 = 0
rate2 = 0
count = 0
wins = 0
for m in range(4,7):
     for d in range(1,31):
        schedules = mlb.games(2021,m,d)
        games = mlb.combine_games(schedules)
        for game in games:
            try:
                pitcher1 = game.w_pitcher
                team1 = game.w_team
                pitcher2 = game.l_pitcher
                team2 = game.l_team
            except AttributeError:
                continue
            p1 = season21.loc[season21['Name'] == pitcher1]
            p1ERA = p1.ERA.values.astype(float)
            p1WHIP = p1.WHIP.values.astype(float)
            p1so9 = p1.SO9.values.astype(float)
            t1 = teambatting.loc[teambatting['Team'] == init[team1]]
            t1BA = (t1.H/t1.AB).values.astype(float)
            t1OBP = t1.OBP.values.astype(float)
            t1SLG = t1.SLG.values.astype(float)
            t1 = teampitching.loc[teambatting['Team'] == init[team1]]
            t1ERA= t1.ERA.values.astype(float)
            t1WHIP = t1.WHIP.values.astype(float)
            t1SO9= (t1.SO*9/t1.IP).values.astype(float)
            p1ERA = normalize(p1ERA, min_era, max_era)
            p1WHIP = normalize(p1WHIP, min_whip, max_whip)
            p1so9 = normalize(p1so9, min_so9, max_so9)
            t1ERA= normalize(t1ERA, min_bera, max_bera)
            t1WHIP = normalize(t1WHIP, min_bwhip, max_bwhip)
            t1SO9= normalize(t1SO9, min_bso9, max_bso9)
            t1BA = normalize(t1BA, min_ba,max_ba)
            t1OBP = normalize(t1OBP, min_obp, max_obp)
            t1SLG = normalize(t1SLG, min_slg, max_slg)
            Spitching = pitching(A , x1, p1ERA, y1, p1WHIP, z1, p1so9)
            bpitching = pitching(B, x2, t1ERA, y2, t1WHIP, z2, t1SO9)
            batting = batting_form(C, x3 , t1BA, y3, t1OBP, z3, t1SLG)
            rate1 = Spitching + bpitching + batting
            ######################
            ######################
            ######################
            p2 = season21.loc[season21['Name'] == pitcher2]
            p2ERA = p2.ERA.values.astype(float)
            p2WHIP = p2.WHIP.values.astype(float)
            p2so9 = p2.SO9.values.astype(float)
            t2 = teambatting.loc[teambatting['Team'] == init[team2]]
            t2BA = (t2.H/t2.AB).values.astype(float)
            t2OBP = t2.OBP.values.astype(float)
            t2SLG = t2.SLG.values.astype(float)
            t2 = teampitching.loc[teambatting['Team'] == init[team2]]
            t2ERA= t2.ERA.values.astype(float)
            t2WHIP = t2.WHIP.values.astype(float)
            t2SO9= (t2.SO*9/t2.IP).values.astype(float)
            p2ERA = normalize(p2ERA, min_era, max_era)
            p2WHIP = normalize(p2WHIP, min_whip, max_whip)
            p2so9 = normalize(p2so9, min_so9, max_so9)
            t2ERA= normalize(t2ERA, min_bera, max_bera)
            t2WHIP = normalize(t2WHIP, min_bwhip, max_bwhip)
            t2SO9= normalize(t2SO9, min_bso9, max_bso9)
            t2BA = normalize(t2BA, min_ba,max_ba)
            t2OBP = normalize(t2OBP, min_obp, max_obp)
            t2SLG = normalize(t2SLG, min_slg, max_slg)
            Spitching = pitching(A, x1, p2ERA, y1, p2WHIP, z1, p2so9)
            bpitching = pitching(B , x2, t2ERA, y2, t2WHIP, z2, t2SO9)
            batting = batting_form(C, x3 , t2BA, y3, t2OBP, z3, t2SLG)
            rate2 = Spitching + bpitching + batting
            try:
                if rate1[0] > rate2[0]:
                    wins = wins + 1
                count = count + 1
                print(wins/count)
            except IndexError: continue
```
